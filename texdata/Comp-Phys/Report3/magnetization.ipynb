```python
#import系
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
#データのload
up=np.loadtxt("iron_up-dos.dat")
dn=np.loadtxt("iron_dn-dos.dat")
```

```python
print(up)
```

```python
#Fermi Energy(eV)
fermienergy=12.6256
```

```python
plt.title("spin-up-and-dn")
plt.plot(up[:,0],up[:,1],label='spin-up')
plt.plot(dn[:,0],dn[:,1],label='spin-dn')
```

```python
#dataをfermienergyの所で切断する．
#この時，二次元配列を途中でcutしたいので
#配列の初期化は,np.emptyで形を合わせる！
#追加する時には，しっかりと[[]]の形で加えないといけない！

Spinup=np.empty((0,2) )
print(up[0,],up[1,])

print(Spinup)

#array1=up[0,0]
#array2=up[0,1]
#ll=[[array1,array2]]
#print(array)
#Spinup=np.append(Spinup,[up[0,]],axis=0)
#print(Spinup)
#Spinup=np.append(Spinup,[up[1,]],axis=0)

i=0

while up[i,0] < fermienergy :
    Spinup=np.append(Spinup,[up[i,]],axis=0)
    i=i+1
 
print(Spinup) 
```

```python
Spindn=np.empty((0,2) )
print(dn[0,],dn[1,])

print(Spindn)

i=0

while dn[i,0] < fermienergy :
    Spindn=np.append(Spindn,[dn[i,]],axis=0)
    i=i+1
 
print(Spindn) 

```

```python
plt.title("Spin DOS")
plt.plot(Spinup[:,0],Spinup[:,1])
plt.plot(Spindn[:,0],Spindn[:,1])
```

```python
#DOSの積分で個数を求める
Nup=integrate.simps(Spinup[:,1],Spinup[:,0])
Ndn=integrate.simps(Spindn[:,1],Spindn[:,0])
```

```python
#結果の表示
print('スピンupは',Nup)
print('スピンdnは',Ndn)
```
